```python
import numpy as np

print("请输入评价对象的数量：")
object_num = eval(input())

print("请输入评价指标的数量：")
index_num = eval(input())

print("请输入每个评价指标对应的类型矩阵  1:极大型  2:极小型  3:中间型  4:区间型")
kind = input().split(" ")   # 此时的kind是字符串类型的列表

print("请输入矩阵：")
A = np.zeros((object_num, index_num))
for i in range(object_num):
    A[i] = input().split(" ")       # 此时的A[i]是字符串类型的列表
    A[i] = list(map(float, A[i]))   # 将字符串类型的列表转换为浮点数类型的列表
print("输入矩阵为：\n{A}")

def minToMax(maxA, x):  # 极小型指标转换为极大型指标
    return np.array([[(maxA - i)] for i in x])
   
def midToMax(bestA, x):  # 中间型指标转换为极大型指标
    tmp = [abs(bestA - i) for i in x]
    maxTmp = max(tmp)
    if maxTmp == 0:
        return np.array([[1] for i in x])
    else:
        return np.array([[(maxTmp - i) / maxTmp] for i in tmp])
    
def regToMax(lowA, highA, x):  # 区间型指标转换为极大型指标
    maxTmp = max(lowA - min(x), max(x) - highA)
    if maxTmp == 0:
        return np.array([[1] for i in x])
    else:
        tmp = []
        for i in x:
            if i < lowA:
                tmp.append([(maxTmp - (lowA - i)) / maxTmp])
            elif i > highA:
                tmp.append([(maxTmp - (i - highA)) / maxTmp])
            else:
                tmp.append([1.0])
        return np.array(tmp)

X = np.zeros((object_num, index_num))
for j in range(index_num):
    if kind[j] == "1":      # 极大型指标
        v = np.array(A[:, j])
    elif kind[j] == "2":    # 极小型指标
        maxA = max(A[:, j])
        v = minToMax(maxA, A[:, j])
    elif kind[j] == "3":
        print("中间型指标，请输入最优值：")
        bestA = eval(input())
        v = midToMax(bestA, A[:, j])
    elif kind[j] == "4":
        print("区间型指标，请输入区间[a, b]值a：")
        lowA = eval(input())
        print("区间型指标，请输入区间[a, b]值b：")
        highA = eval(input())
        v = regToMax(lowA, highA, A[:, j])
    
    if j == 0:
        X = v.reshape(-1, 1)
    else:
        X = np.hstack((X, v.reshape(-1, 1)))
print("转换后的矩阵为：\n{X}")

# 归一化处理
X = X.astype(float)
for j in range(index_num):
    X[:, j] = X[:, j] / np.sqrt(sum(X[:, j] ** 2))
print("归一化处理后的矩阵为：\n{X}")

# 计算权重
print("请输入各指标的权重(以空格分隔)：")
w = np.array(list(map(float, input().split(" "))))
print("权重向量为：\n{w}")

# 最大最小值距离的计算
x_max = np.max(X, axis=0)
x_min = np.min(X, axis=0)
d_z = np.sqrt(np.sum(w * np.square(X - np.tile(x_max, (object_num, 1))), axis=1))
d_f = np.sqrt(np.sum(w * np.square(X - np.tile(x_min, (object_num, 1))), axis=1))
print("各评价对象的最大值距离为：\n{d_z}")
print("各评价对象的最小值距离为：\n{d_f}")

# 计算最终得分
s = d_f / (d_f + d_z)
for i in range(len(s)):
    print("第{}个评价对象的得分为：{}".format(i + 1, s[i]))
# 90 4 35 15
# 85 3 50 21
# 92 4.5 40 19
# 88 3.5 38 17

# 40 16 18
# 0.25 0.25 0.25 0.25
```
